# Disaster Tweet

```python
import pandas as pd
import numpy as np
from textblob import TextBlob
PATH = 'data/'
df = pd.read_csv(PATH + 'data.csv')
df = df.sample(df.shape[0], random_state=42)
#df.text = df.text_clean.astype(str)
df = df[['id', 'text', 'target']]
text = df[df.target == 1].sample(20).text.iloc[1]
print(text)
```

```python
import spacy
nlp = spacy.load("en_core_web_lg")
doc = nlp(text)
```

```python
for token in doc:
    print(token, token.pos_)
```

```python
for ent in doc.ents:
    print(ent, ent.label_)
```

```python
import spacy
from spacy import displacy
displacy.serve(doc, style="dep")
```

```python
def CS(a, b):
    from sklearn.metrics.pairwise import cosine_similarity
    return cosine_similarity(a.reshape(1, -1), b.reshape(1, -1)).item()
```

```python
for token1 in doc:
    for token2 in doc:
        print(token1, token2, CS(token1.vector, token2.vector))
```
